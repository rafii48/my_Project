```python
tickers = ['RBC','CIBC','BMO','Suncor','Encana','CNQ']
bo = pd.concat([rbc,cibc,bmo,suncor,encana,cnq],axis=1,keys=tickers)
bo.head()
```

```python
bo.columns.names
```

```python
bo.columns.names = ['Entity','Stock']
bo.columns.names
```

```python
bo.head()
```

```python
max_stock= bo.xs(key='Close',axis=1,level='Stock').max()
max_stock
```

```python
min_stock= bo.xs(key='Close',axis=1,level='Stock').min()
min_stock
```

```python
fig, axes= plt.subplots(ncols=2,figsize=(10,4))
sns.barplot(x=max_stock.index, y=max_stock.values,ax=axes[0])
axes[0].set_ylim(0,120)
axes[0].set_ylabel('max_stock')

sns.barplot(x=min_stock.index, y=min_stock.values,ax=axes[1])
axes[1].set_ylim(0,120)
axes[1].set_ylabel('min_stock')
```

```python
for tick in tickers:
    bo[tick]['Close'].plot(figsize=(10,6),label=tick)
plt.legend()
```

```python
bo.xs(key='Close', axis=1,level='Stock').iplot()
```
