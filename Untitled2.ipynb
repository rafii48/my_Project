```python
import scipy

import random
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
file = 'DHS_39BB.las'
```

```python
data=np.loadtxt(file,skiprows=2)
mnemonic =['depth','calp','pef','rhob','dho','gr','npls','rshal','rdeep','rmic','vs','vp']


data = pd.DataFrame(data,columns=mnemonic)
```

```python
data.head()
```

```python
depth = data['depth']
calp = data['calp']
pef = data['pef']
rhob = data['rhob']
dho = data['dho']
gr = data['gr']
npls = data['npls']
rshal = data['rshal']
rdeep = data['rdeep']
rmic = data['rmic']
vs = data['vs']
vp = data['vp']
```

```python
vp=304.8/data.vp
vs=304.8/data.vs
dens=data.rhob+data.dho

data=data.assign(P=vp,Vs=vs,dens=dens)
data.head()

#bad caliper reading
wo=data[(data['calp']<=13)]
```

```python
#log gr transformation
IGR=(data.gr-min(wo['gr']))/(max(wo['gr'])-min(wo['gr']))
VCL=(data.gr-min(wo['gr']))/(210-min(wo['gr']))

#adding new column in data
data = data.assign(IGR=IGR)
data = data.assign(VCL=VCL)

grmin=data.loc[data['VCL']<=0.12,'gr'].iloc[0]
grmax=data.loc[data['VCL']>0.3,'gr'].iloc[0]
vsh=(data.gr-grmin)/(grmax-grmin)


data=data.assign(vsh=vsh)

#log poro
psh=2.76+0.001*((data.dens-2)-230*np.exp(-4*(data.dens-2)))
pma=2.65*(1-vsh)+psh*vsh
por=(pma-data.dens)/(pma-1.01)

data = data.assign(por=por)
data.head()
```

```python
#rock physics calculation
VPVS = data.vp/data.vs #create vp/vs logs
AI = data['P']*data['dens']
pr = 0.5*(((VPVS)**2)-2)/(((VPVS)**2)-1) #Poisson's Ratio
myu = (vs*vs)*data['dens'] #Bulk Modulus (GPa)
young = 2*myu*(1+pr) # Young's Modulus (GPa)
bulk = data['dens']*(vp*vp)-(4/3)*myu # Bulk Modulus (GPa)
m = (vp**2)*data['dens'] #compressional modulus (GPa)

SI = vs*data['dens']
LR = SI**2
MR = AI**2 - 2*SI**2


data= data.assign(VPVS=VPVS, P=vp, VS=vs,AI=AI,PR=pr, G=myu, K=bulk, M=m, E=young,SI=SI,LR=LR,MR=MR)
data.head()
```

```python
Kmix=37
Mmix=44
Mb=2.6
#rock physics model voigt-reuss-hill
frac=pd.Series(np.linspace(0,1,100))
kv=((1-frac)*Kmix+frac*2.6)
mv=(1-frac)*Mmix
kr=1/((1-frac)/Kmix+frac/2.6)
mr=(1-frac)/Mmix
kh=(kv+kr)/2
mh=(mv+mr)/2

#rock physics model hertz-mindzin
kp=37+((frac)/((1/(2.6-37))+((1-frac)/(37+(4*Mmix/3)))))
kn=2.6+((1-frac)/((1/(37-2.6))+((frac)/(2.6+(4*0/3)))))
#lithology classification
cs=data[(data['vsh']<=0.12)&(data['calp']<=14)&(data['P']<=4)]
ss=data[(data['vsh']>0.3)&(data['calp']<=14)&(data['P']<=4)]
sh=data[(data['vsh']>=1)&(data['calp']<=14)&(data['P']<=4)]
ls=data[(data['P']>=4)]
```

```python
plt.scatter(cs['por'],cs['K'],c='green')
plt.scatter(ss['por'],ss['K'],c='lime')
plt.plot(frac,kv,c='black')
plt.plot(frac,kr,c='blue')
plt.plot(frac,kh,c='gold')
plt.plot(frac,kp,c='green')     
plt.plot(frac,kn,c='red')
plt.ylim(0,40)
plt.legend(['voigt','reuss','hill','HM+','HM-','clean sand','shaly sand','shale'])
plt.title('Rock Physics Model')
plt.xlabel('Porosity')
plt.ylabel('Bulk Modulus (Gpa)')
plt.grid()
```
